```python
prefix = "../raw_data/data_Track/Track1000_" 
num = np.arange(1,1001) # np.arange(101)[1:]
suffix = ".csv"
```

```python
allfiles = list(map(lambda x: prefix + str(x) + suffix, num))
allfiles
```

```python
sqlite3.register_adapter(np.int64, lambda val: int(val))
sqlite3.register_adapter(np.int32, lambda val: int(val))
```

```python
sql = sqlite3.connect('../raw_data/spotify_working.db')
cur = sql.cursor()

cur.execute('''CREATE TABLE IF NOT EXISTS track
            (Playlistid BIGINT, Trackid BIGINT, Artist_Name , Track_Name, Album_Name, Track_Duration BIGINT, Artist_uri,Track_uri,Album_uri)''') # create the table if it doesn't already exist
```

```python
for inx, file in enumerate(allfiles):
    trial = pd.read_csv(file)
    trial = trial.drop(["Unnamed: 0"], axis=1)
    print(file)
    
    cur.executemany('''INSERT INTO track
        (Playlistid, Trackid, Artist_Name, Track_Name, Album_Name, Track_Duration, Artist_uri,Track_uri,Album_uri)
        VALUES (?, ?, ?, ?, ?,?,?,?,?)''', trial.values.tolist())
```

```python
# f.close()
sql.commit()
sql.close()
```

# Check by query the SQLite database

```python
def viz_tables(cols, query):
    q = cursor.execute(query).fetchall()
    framelist = dict()
    for i, col_name in enumerate(cols):
        framelist[col_name] = [col[i] for col in q]
    return pd.DataFrame.from_dict(framelist)
```
